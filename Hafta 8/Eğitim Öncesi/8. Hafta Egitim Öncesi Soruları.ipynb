# Problem 1 
Kullanıcıdan aldığı ürün bilgisini (ad, fiyat) urunler.txt dosyasına kayıt eden fonksiyonu yazınız.

````python
def kaydetme(ad, fiyat):
    with open("urun_bilgi.txt", "a", encoding="utf-8") as file:
        file.write(f"{ad}, {fiyat}\n")
        file.close()
ad1 = input("Ad?")
fiyat1 = int(input("Fiyat?"))
kaydetme(ad1,fiyat1)
````
