# Problem 1
Vehicle sınıfından miras alan bir Bus sınıfı oluşturun. Bus.seating_capacity()'nin kapasite bağımsız değişkenine varsayılan olarak 50 değerini verin.

```
Output:
> The seating capacity of a bus is 50 passengers
```


````python
class Vehicle:
    seating_capacity = 50
    def __init__ (self, name, mass, max_speed, seating_capacity):
        self.name = name
        self.mass = mass
        self.max_speed = max_speed
        self.seating_capacity = seating_capacity
    def show(self):
        print(self.name, self.mass, self.max_speed, self.seating_capacity)
    def ayar(self, sayi):
        self.seating_capacity = sayi
class Bus(Vehicle):
    def __init__ (self, name, mass, max_speed, seating_capacity):
        super().__init__(name, mass, max_speed, seating_capacity)

    def show(self):
        print(self.name, self.mass, self.max_speed, self.seating_capacity)
otobus = Bus("Abbas", 1000, 120, 77)
otobus.ayar(79)
otobus.show()
````
